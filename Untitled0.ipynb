```python
base_url = "https://en.wikipedia.org"
url = "https://en.wikipedia.org/wiki/1992_in_film"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
movie_links = []
for link in soup.select("a[href^='/wiki/'][href*='_film']"):
    movie_links.append(base_url + link['href'])
```

```python
movie_links = list(set(movie_links))[:40]
```

```python
import os
```

```python
output_dir = "dataset/1992_movies"
os.makedirs(output_dir, exist_ok=True)
```

```python
for i, link in enumerate(movie_links):
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    text = soup.get_text(separator='\n')
    with open(f"{output_dir}/movie_{i+1}.txt", "w", encoding="utf-8") as f:
        f.write(text)
```

```python
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
files = os.listdir(output_dir)
stop_words = set(stopwords.words('english'))
```

```python
import nltk

nltk.download('stopwords')
```

```python
files = os.listdir(output_dir)
stop_words = set(stopwords.words('english'))
documents = []
for file in files:
    with open(f"{output_dir}/{file}", "r", encoding="utf-8") as f:
        text = f.read()
        tokens = word_tokenize(text.lower())
        tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
        documents.append(" ".join(tokens))
```

```python
nltk.download('punkt_tab')
```

```python
from collections import Counter
```

```python
for doc in documents:
    word_counts = Counter(doc.split())
    print(word_counts.most_common(10))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text = " ".join(documents)  # Combine all movie texts
wordcloud = WordCloud(width=800, height=400).generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)  # Limit features for clarity
tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
from sklearn.cluster import KMeans
```

```python
num_clusters = 10  # Adjust based on results
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(tfidf_matrix)
```

```python
import numpy as np
```

```python
terms = vectorizer.get_feature_names_out()
for i in range(num_clusters):
    cluster_terms = np.argsort(kmeans.cluster_centers_[i])[::-1][:10]
    print(f"Cluster {i+1}: {[terms[idx] for idx in cluster_terms]}")
```

```python
from sklearn.decomposition import PCA
```

```python
import matplotlib.pyplot as plt
```

```python
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(tfidf_matrix.toarray())
```

```python
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=clusters, cmap='viridis')
plt.title("Movie Clusters")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```
